# Simple Linear Regression : Experience vs. Salary

## Objective:
### Build a simple Linear Regression model to predict salary based on years of experience.

## Dataset:

### This dataset contains information on salaries and years of experience for individuals working in various professions. It is designed for exploring and building simple linear regression models to understand and predict salary based on years of experience.
### Link to the dataset: https://www.kaggle.com/datasets/karthickveerakumar/salary-data-simple-linear-regression?select=Salary_Data.csv

### Import necessary libraries

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

### Load the dataset

```python
data = pd.read_csv('Salary_Data.csv')
```

```python
data.info()
```

```python
data.shape
```

#### The dataset has only 30 records.

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Define the feature and target variable
x = data['YearsExperience'].values.reshape(-1, 1)   # Feature
y = data['Salary']             # Target variable
```

### Scatter plot

```python
plt.scatter(x, y, color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter plot of Years of Experience vs Salary')
plt.show()
```

#### The points cluster near a line with a positive slope, indicating a strong positive correlation between 'Years of Experience' and 'Salary' in the scatter plot.

#### Simple linear regression for the salary data involves modeling the relationship between an employee's years of experience (independent variable) and their salary (dependent variable). The goal is to find a straight-line equation that best predicts salary based on years of experience. By fitting the model to the data, we can predict salaries for various years of experience and assess how strongly experience influences salary.

## Spliting the data and Training the Model

```python
# Split the data into training and testing sets.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initialize and fit the linear regression model using the training data.
model = LinearRegression()
model.fit(x_train, y_train)

#  Use the trained model to make predictions on the test set.
y_pred = model.predict(x_test)
```

## Model Evaluation

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

 An R-squared value of 0.9024 suggests that the model is doing quite well in explaining the variation in salary based on years of experience.
 The Mean Squared Error (MSE) measures the average squared difference between the observed actual values and the predicted values. A lower MSE indicates that the predictions are closer to the actual values. In this case, the MSE is quite large, indicating that there are still some significant differences between predicted and actual salaries, though MSE can appear large when the scale of the target variable (Salary) is large.

## Residual plot

```python
plt.scatter(x, y - model.predict(x))
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Actual Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

### The randomly scattered residuals suggest that the simple linear regression model is a good fit for this dataset.

## Test Values and Predicted Values

```python
plt.scatter(x_test, y_test, color='green', label='Actual Test Values')  # Plot the actual test values
plt.scatter(x_test, y_pred, color='red', label='Predicted Values')  # Plot the predicted values
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Actual vs Predicted Salary')
plt.legend()
plt.show()
```

```python
# Get the coefficient (slope) and intercept
slope = model.coef_[0]     # 1
intercept = model.intercept_  # 0

# Print the equation of the regression line
print(f"The equation of the regression line is: y = {intercept} + {slope} * x")
```
